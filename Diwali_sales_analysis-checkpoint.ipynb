```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape') 
# here we store data file in variable df
#to avoid encoding error use unicode_escape
```

```python
df.shape #shows the number of rows and columns.
```

```python
df.head() #shows top5 rows and name of columns.
```

```python
df.info()
```

From above we can clearly see that there are two columns named Status and unnamed1 have no value or record in them so they are 
useless to us. We can remove them by using drop method and this process is called data cleaning.

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
#axis=1-->to consider whole vertical column
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values and if we use 'inplace=True' that means that changes took place in original file
df.dropna(inplace=True)
```

*to save changes in original data
1. df.method_name(inplace=True)
2. df=df.method_name()

```python
df.shape
```

```python
# to change datatype
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#rename column
df.rename(columns={'Marital_Status':'Shaadi'})
```

*Analysis on Gender base

```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
#to get values on bar we use for loop
for bars in ax.containers:
    ax.bar_label(bars)
```

here  we can clearly see that maximum orders are placed by females

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

Here we get amount on the basis of groups of  gender. 

To get groups in sorted manner we can use -->sort_values() 

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount', data=sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

*Analysis on the basis of age of customers

```python
df.columns
```

```python
sns.countplot(data=df, x= 'Age Group')
```

```python
# to find sale count on the basis of age as well as on the basis of gender we can use 'hue'
ax=sns.countplot(data=df, x= 'Age Group', hue='Gender')
# for values
for bars in ax.containers:
    ax.bar_label(bars)
```

This bar chart shows that maximum shopping is done by female of age group 26-35.

```python
# Total spending of amount on shopping according to  age group
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

From this we can clearly see that most of the buyers spend money on shopping are between age group of 26-35. 

* Analysis of sale by using state column

```python
# Orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

plt.figure(figsize=(18, 10))
sns.barplot(data=sales_state, x='State', y='Orders')
plt.title("Orders from top 10 states", fontsize=20)
plt.xlabel("States", fontsize=20)
plt.ylabel("Orders", fontsize=20)
```

```python
amount_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.figure(figsize=(25, 10))
sns.barplot(data=amount_state, x='State', y='Amount')
plt.title("Orders from top 10 states", fontsize=20)
plt.xlabel("States", fontsize=20)
plt.ylabel("Amount spend", fontsize=20)
```

Most of the ordres and total spend amount on shopping are from Uttar Pradesh, Maharashtra and Karnataka respectively.
Minimum orders are from Telangana.

* Marital status

```python
ax=sns.countplot(data=df, x= "Marital_Status")

plt.figure(figsize=(10,10))
for bars in ax.containers:
    ax.bar_label(bars)
```

Most shopping is done by married people.

```python
sales_state= df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# another way to set figure size by seaborn library
sns.set(rc={'figure.figsize':[6,5]})
sns.barplot(data= sales_state, x='Marital_Status', y="Amount", hue='Gender')
```

From abovr graphs we can see that most of the buyers are married women and they have high purchasing power.

* Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

From above it can be seen that most orders are given by people who works in IT sector.

```python
sales_occu=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_occu, x='Occupation',y='Amount')
```

Above chart shows that on the basis of occupation whoes purchasing power is high. From chart its looks IT sector people spend more money on shopping.

*Product Category

```python
sns.set(rc={'figure.figsize':(25,10)})
ax= sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

This shows that people like to shop clothing and apparel, food, Electronic and gadgets more.

```python
sales_pro= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,10)})
ax= sns.barplot(data=sales_pro,x='Product_Category',y='Amount')

```

From above chart we can see that most of the spendings on products are from food,clothing and electronics category.

```python
# To see top saling products
top_sale_prod=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=top_sale_prod, x='Product_ID',y='Orders')
```

Conclusion:

Married Women of age group 26-35 years from UP, Maharashtra and Karnataka working in IT sector, Healthcare and Aviation are more likely buy products from Food, Clothing and Electronics Category.
